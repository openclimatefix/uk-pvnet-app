```python
from ocf_data_sampler.select.geospatial import osgb_to_lon_lat
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("uk_gsp_locations_20250109.csv", index_col="gsp_id")
```

```python
df.x_osgb
```

```python
lon, lat = osgb_to_lon_lat(df["x_osgb"].values, df["y_osgb"].values)
```

```python
df["longitude"] = lon
df["latitude"] = lat
```

```python
df = df.drop(["geometry", "x_osgb", "y_osgb"], axis=1)
```

```python
df.to_csv("uk_gsp_locations2_20250109.csv")
```
